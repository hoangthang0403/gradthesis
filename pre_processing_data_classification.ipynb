```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Đọc dữ liệu
tra = pd.read_csv('data/TRA.csv', index_col=0, parse_dates=True)
vnindex = pd.read_csv('data/VNINDEX.csv', index_col=0, parse_dates=True)
pharma = pd.read_csv('data/pharma.csv', index_col=0, parse_dates=True)

# Giữ lại cột 'close'
tra = tra[['close']].rename(columns={'close': 'TRA'})
tra.index = pd.to_datetime(tra.index)

vnindex = vnindex[['time','close']]
vnindex.set_index('time', inplace=True)
vnindex.index = pd.to_datetime(vnindex.index)
vnindex.drop_duplicates(inplace=True)
vnindex = vnindex[['close']].rename(columns={'close': 'VNINDEX'})
vnindex

pharma = pharma[['close']].rename(columns={'close': 'PHARMA'})
pharma.index = pd.to_datetime(pharma.index)
pharma

df = pd.concat([tra,pharma,vnindex], axis=1,join='inner')

# Đọc dữ liệu
df.index = pd.to_datetime(df.index)

df['target'] = (df['TRA'].shift(-1) > df['TRA']).astype(int)

df['VNINDEX_lag2'] = df['VNINDEX'].shift(2)
df['PHARMA_lag1'] = df['PHARMA'].shift(1)

df.dropna(inplace=True)

df.drop(['PHARMA','VNINDEX'],axis=1,inplace=True)


df.to_csv('data/TRA_PHARM_VNINDEX.csv')

df
```
